```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['MURDER','ATTEMPT TO MURDER']])
y_predicted
df['cluster'] = y_predicted
df.drop('cluster',axis='columns')
df
```

```python
dataset1 = df[df.cluster==0]
dataset2 = df[df.cluster==1]
dataset3 = df[df.cluster==2]
plt.scatter(dataset1.MURDER,dataset1['ATTEMPT TO MURDER'],color='red',label='Points')
plt.scatter(dataset2.MURDER,dataset2['ATTEMPT TO MURDER'],color='green',label='Points')
plt.scatter(dataset3.MURDER,dataset3['ATTEMPT TO MURDER'],color='black',label='Points')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='Centroid')
plt.title("MURDER Vs ATTEMPT TO MURDER OF INDIA ON 2012")
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.legend()
```

```python
k_rng =range(1,10)
sse = []
for k in k_rng:
    km =KMeans(n_clusters=k)
    km.fit(df[['MURDER','ATTEMPT TO MURDER']])
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
